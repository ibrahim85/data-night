# Demo: Moscow CCTV cameras

## Imports

```python
import pandas
import pygeoplot as gp
from IPython.display import Image
```

## Download the data

#### Let we download [the dataset](http://data.mos.ru/datasets/1498) of outdoor CCTV cameras in Moscow from [Moscow Open Data Portal](http://data.mos.ru/).

```python
!wget http://data.mos.ru/datasets/download/1498/1/1 -O moscow_traffic_cameras.csv
```

#### It is just a CSV file in Excel dialect:

```python
!head -n 3 moscow_traffic_cameras.csv
```

#### So we read it into [pandas](http://pandas.pydata.org/) data frame:

```python
data = pandas.read_csv('./moscow_traffic_cameras.csv', sep=';', index_col=0, usecols=range(10))
print 'Number of rows:', data.shape[0]
data.head()
```

## Show cameras on the map

```python
map_center = list(data.ix[:10, ['Y_WGS84', 'X_WGS84']].mean())

cameras_map = gp.Map()
gp.placemarks_from_df(cameras_map, data.head(n=10), lat_col='Y_WGS84', lng_col='X_WGS84')
cameras_map.set_state(map_center, 13)
cameras_map.display()
```

#### Each landmark corresponds to the row in our data frame. Hover the landmark to see its index in the data frame, click it to see the corresponding row:

```python
Image('screen-click-geo-dataframe.png')
```

## Show heatmap of surveillance

```python
map_center = list(data.ix[:, ['Y_WGS84', 'X_WGS84']].median())
heatmap_points = [list(point) for i, point in data.ix[:, ['Y_WGS84', 'X_WGS84']].iterrows()]

camera_density = gp.Map()
camera_density.add_heatmap(heatmap_points)
camera_density.set_state(map_center, 9)
camera_density.display(width=1000)
```

## Saving visualization to the stand-alone HTML file

```python
camera_density.save_html('demo-moscow-cctv-heatmap.html')
```
